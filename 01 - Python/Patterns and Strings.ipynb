```python
#Prime or not
n = int(input())
if n<=1:
    print("Not a prime")
for i in range(2, n+1):
    if n%i == 0:
        print("Not Prime")
        break
    print("Prime")
```

```python
#Palindrome of Integers
a = 12321
temp = a
reverse = 0
while a>0:
    digit = a % 10
    reverse = reverse*10 + digit
    a = a//10


if reverse == temp:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```
