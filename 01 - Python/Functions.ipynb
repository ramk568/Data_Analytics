```python
#max 3 numbers 
def maz_num(a,b,c):
    if a>b and b>c:
        print("a is greater")
    elif b>a and b>c:
        print("b is greater")
    else:
        print("c is greatest")
maz_num(1,2,3)
```

```python
#check prime
def check_prime(n):
    if n==1:
        print("is not prime")
    if n==2:
        print("is a prime")
    if n>3:    
        for i in range(2,n+1):
            if n%i==0:
                print("is not a prime")
                break
            else:
                print("is a prime")
check_prime(4)
    
```

```python
def add(n):
    total = 0
    for i in n:
        total+=1
    return(total)
print(add([1,2,3,4]))
```

```python
def add(n):
    if len(n)==1:
        return (n[0])
    else:
        return (n[0]+add(n[1:]))
print(add([12,3,4,5]))
```

```python
#fibonacci using recursion
def fibo(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
print(fibo(9))
```
